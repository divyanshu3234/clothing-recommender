```python

import cv2
import mediapipe as mp

# Initialize Mediapipe Pose model
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()

def detect_landmarks(image_path):
    # Read the image
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Process the image to find body landmarks
    results = pose.process(image_rgb)
    if results.pose_landmarks:
        return results.pose_landmarks.landmark
    else:
        print("No landmarks detected.")
        return None

```

```python
import math

def calculate_distance(point1, point2):
    # Calculate Euclidean distance between two points
    return math.sqrt((point2.x - point1.x) ** 2 + (point2.y - point1.y) ** 2)

# def extract_measurements(landmarks):
#     # Define key points
#     shoulder_left = landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER]
#     shoulder_right = landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER]
#     hip_left = landmarks[mp_pose.PoseLandmark.LEFT_HIP]
#     hip_right = landmarks[mp_pose.PoseLandmark.RIGHT_HIP]
#     # Calculate measurements
#     shoulder_width = calculate_distance(shoulder_left, shoulder_right)
#     hip_width = calculate_distance(hip_left, hip_right)
#     print("shoulder_width : ", shoulder_width,"\n")
#     print("hip_width : ", hip_width,"\n")
#     return {"shoulder_width": shoulder_width, "hip_width": hip_width}


def extract_measurements(landmarks):
    # Define key points
    shoulder_left = landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER]
    shoulder_right = landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER]
    hip_left = landmarks[mp_pose.PoseLandmark.LEFT_HIP]
    hip_right = landmarks[mp_pose.PoseLandmark.RIGHT_HIP]

    #Defining hamari height :
    head = landmarks[mp_pose.PoseLandmark.NOSE]
    foot = landmarks[mp_pose.PoseLandmark.LEFT_FOOT_INDEX]

    height = calculate_distance(head, foot)
    actualheight=163
    factor = actualheight/height
    print("factor : ", factor,"\n")
    print("height : ", height,"\n")
    print("height in cm : ", (height*factor)+10,"\n")

    # Calculate measurements
    shoulder_width = calculate_distance(shoulder_left, shoulder_right)
    hip_width = calculate_distance(hip_left, hip_right)
    print("shoulder_width : ", (shoulder_width*factor)-15,"\n")
    print("hip_width : ", (hip_width*factor)-15,"\n")

    return {"shoulder_width": shoulder_width, "hip_width": hip_width}
```

```python
def scale_measurements(measurements, reference_real_world, reference_pixel):
    scale_factor = reference_real_world / reference_pixel
    for key in measurements:
        measurements[key] *= scale_factor
    return measurements
```

#Importing & Creating A CSV

```python
def get_body_measurements(image_path, reference_real_world, reference_pixel):
    landmarks = detect_landmarks(image_path)
    if landmarks:
        measurements = extract_measurements(landmarks)
        measurements_scaled = scale_measurements(measurements, reference_real_world, reference_pixel)
        return measurements_scaled
    return None

# Example usage
image_path = "akclose.jpeg"
reference_real_world = 32  # Known width in cm
reference_pixel = 0.32     # Measured pixel width between shoulders

measurements = get_body_measurements(image_path, reference_real_world, reference_pixel)
print("Body measurements (in cm):", measurements)
```
